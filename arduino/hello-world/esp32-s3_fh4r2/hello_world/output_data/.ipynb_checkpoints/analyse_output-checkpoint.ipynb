hello_world.ino prints values from the hello_world model for one complete cycle  
The output has the shape:  
`x_value: 0.000000, y_value: 0.000000`  
There is some more text in the file, which we simply skip. We should find 1000 x,y value pairs  
The output from the arduino program has been captured with:  
tio /dev/ttyACM0 | tee sin.txt  
Type this command and then push the reset button on the micro-controller.  
Once you see that the output has stopped (even though it seems that it has not finished),
exit tio with Ctrl t q 

Your text file should look like this:  
![sin.txt](../images/hello_world_results.png "hello_world result file")

```python
import matplotlib.pyplot as plt
from math import sin, pi
```

Read only lines that have the text x_value in them

```python
helloWorldDataFilename = "sin.txt"
dataFile = open(helloWorldDataFilename,"r")
x_values = []
y_values = []
fileContent = []
line = dataFile.readline()
while line:
    start = line.find("x_value")
    if start != -1:
        line = line[start:]
        fileContent.append(line.strip())
    line = dataFile.readline()
```

```python
noOfDataPoints = len(fileContent)
print(noOfDataPoints)
```

```python
for i in range(20):
    print(fileContent[i])
```

get the first line, split it on the ',' and extract the numbers as strings

```python
values = fileContent[0].split(',')
print("What is to become the x value: ",values[0]);
print("What is to become the y value:",values[1]);  
```

Remove everthing before and including the ':' character

```python
x_string = values[0][values[0].find(':') +1 :]
y_string = values[1][values[1].find(':') +1 :]
print("x string: ",x_string)
print("y string: ",y_string)
```

Convert the strings to float

```python
x = float(x_string)
y = float(y_string)
print("type of x,y: ",type(x))
print("values: x: {:f}, y: {:f}".format(x,y))
```

Go through one cycle and extract the x and y values

```python
for index in range(noOfDataPoints):
    values = fileContent[index].split(',')
    x_values.append(float(values[0][values[0].find(':') +1 :]))
    y_values.append(float(values[1][values[1].find(':') +1 :]))
```

Plot the results and compare them to a regular sine cycle

```python
x_comparison = []
y_comparison = []
for i in range(noOfDataPoints):
    x_comparison.append(2.0*pi*i/noOfDataPoints)
    y_comparison.append(sin(x_comparison[i]))
```

```python
for i in range(20):
    print(x_comparison[i],y_comparison[i])
```

```python
plt.title("Output from the hello world Tensorflow model")
plt.plot(x_values,y_values)
plt.plot(x_comparison,y_comparison)
```
