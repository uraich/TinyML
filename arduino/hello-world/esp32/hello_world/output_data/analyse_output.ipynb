## Reads the output from the hello world program and plots it
We restrict ourselves to a single cycle 

```python
import matplotlib.pyplot as plt
```

Read the data from the file line by line

```python
helloWorldDataFilename = "output.txt"
dataFile = open(helloWorldDataFilename,"r")
x_values = []
y_values = []
fileContent = dataFile.readlines()
```

```python
print(fileContent[0:5])
```

Find the first line where the x value is zero

```python
first_index = 0
while not "0.0000" in fileContent[first_index]:
    first_index += 1
print("Index of line starting a sine cycle: ",first_index)
print("Line with first value: ",fileContent[first_index])
```

```python
last_index = first_index+1
while not "0.0000" in fileContent[last_index]:
    last_index += 1
print("Index of line starting a sine cycle: ",last_index)
print("Line starting a new cycle: ",fileContent[last_index])
```

get the first line, split it on the ',' and extract the numbers as strings

```python
index = first_index
values = fileContent[first_index].split(',')
print("What is to become the x value: ",values[0]);
print("What is to become the y value: ",values[1]);  
```

Remove everthing before and including the ':' character

```python
x_string = values[0][values[0].find(':') +1 :]
y_string = values[1][values[1].find(':') +1 :]
print("x string: ",x_string)
print("y string: ",y_string)
```

Convert the strings to float

```python
x = float(x_string)
y = float(y_string)
print("type of x,y: ",type(x))
print("values: x: {:f}, y: {:f}".format(x,y))
```

Go through one cycle and extract the x and y values

```python
for index in range(first_index,last_index):
    values = fileContent[index].split(',')
    x_values.append(float(values[0][values[0].find(':') +1 :]))
    y_values.append(float(values[1][values[1].find(':') +1 :]))
```

```python
plt.title("Output from the hello world Tensorflow model")
plt.plot(x_values,y_values)
```
