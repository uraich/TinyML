# Read the example wav file
First we decode the file header, then we get at the audio data

```python
no_data = open('no-example.wav','rb')
no_buffer = no_data.read()
no_data.close()
```

```python
print("Show the wav header information")
riff = no_buffer[:4].decode()
print("File type: {}".format(riff))
```

```python
file_size = no_buffer[7] << 24 | no_buffer[6] << 16 | no_buffer[5] << 8 | no_buffer[4] 
print("File size: 0x{:d}".format(file_size))
```

```python
audio_type = no_buffer[8:12].decode()
print("Audio type: {}".format(audio_type))
```

```python
format_marker = no_buffer[12:15].decode()
print("Format marker: {}".format(format_marker))
```

```python
data_length = no_buffer[17] << 8 | no_buffer[16]
print("Data length in bits: {:d}".format(data_length))
```

```python
type_format = no_buffer[21] << 8 | no_buffer[20]
if type_format == 1:
    print("PCM - 2 byte integer")
else:
    print("Unknown format: {d}".format(type_format))
```

```python
no_of_channels = no_buffer[23] << 8 | no_buffer[22]
print("No of channels: {:d}".format(no_of_channels))
```

```python
sample_rate = no_buffer[27] << 24 | no_buffer[26] << 16 | no_buffer[25] << 8 | no_buffer[24]
print("Sample rate: {:d} Hz".format(sample_rate))
```

```python
s_rate_bps_ch = no_buffer[31] << 24 | no_buffer[30] << 16 | no_buffer[29] << 8 | no_buffer[28]
print("(Sample rate * Bits per sample * Channels)/8: {:d}".format(s_rate_bps_ch))
```

```python
bts_ch = no_buffer[35] << 8 | no_buffer[34]
print("(Bits per sample * channels)/8: {:d}".format(bts_ch))
```

```python
bits_per_sample = no_buffer[35] << 8 | no_buffer[34]
print("Bits per sample: {:d}".format(bits_per_sample))
```

```python
data_section = no_buffer[36:40].decode()
print("Start of data section: {}".format(data_section))
```

```python
data_section_length = no_buffer[43] << 24 | no_buffer[42] << 16 | no_buffer[41] << 8 | no_buffer[40]
print("Length of data section: {:d}".format(data_section_length))
```
