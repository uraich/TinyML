## Create a bmp file from grayscale pixel data 

We take a grayscale image, taken from the esp32 camera and saved onto a file, and integrate the pixel data into a bmp file
The program demonstrates, how to create a bmp (bitmap) file from grayscale data, taken from the esp32 camera.
The program will later be implemented in Python code onn the esp32s3-fn8_cam and used in a web server streamin the data to a browser.
Finally it will be integrated into the person detection TinyML code. This allows to see the image which is given to the person detection neural network model

```python
import struct
from PIL import Image
```

```python
BI_RGB = 0 # no compression
```

```python
def printByteArray(buffer):
    for i in range(len(buffer) -1):
        print("0x{:02x}, ".format(buffer[i]),end="")
    print("0x{:02x}".format(buffer[len(buffer) -1]))
```

### Create the info header

The info header has this format:  
The offset i calculated from the beginning of the bmp file.  
    
| signification | value | offset | size in bytes |
| --- | --- | --- | --- |
| header size |  | 14 | 4|
| bitmap width in pixels |  | 18 | 4 |
| bitmap height in pixels |  | 22 | 4 |
| no of colors | 1 | 26 | 2 |
| number of bits per pixel |  | 28 | 2 |
| compression method |  | 30 | 4 |
| image size |  |  34 | 4 |
| horizontal resolution pixels/m |  | 38 | 4 |
| vertical resolution pixels/m |  | 42 | 4 |
| number of colors in palette |  | 46 | 4 |
| number of important colors |  | 50 | 4 |

```python
infoHeader = bytearray(40)
infoHdrSize = len(infoHeader)
imageWidth = 96
imageHeight = 96
noOfPlanes = 1
noOfBitsPerPixel = 8
compression = BI_RGB                    # no compression
imageSize = 0                           # this works for uncompressed images
imageSize = imageWidth * imageHeight * 2
hor_resolution = 0                      # no preference
ver_resolution = 0                      # no preference
noOfColors = 256                        # only used with color maps
importantColors = 0
infoHdr = struct.pack("3i2h6i",infoHdrSize,imageWidth,imageHeight,noOfPlanes,
                      noOfBitsPerPixel,compression,imageSize,hor_resolution,ver_resolution,
                      noOfColors,importantColors)
```

```python
print("Length of the info header: ",len(infoHeader))
```

### Create the color palette

It consists of 256 32 bit values with A: alpha, R: red, G: green,B: blue  
AARRGGBB  
AA is always 0xff which means: fully opaque  
RR, GG, BB have always the same values (resulting in a gray color) starting from 0: black to 0xff: white

```python
palette = bytearray(256*4)
for i in range(256):
    palette[4*i]   = i
    palette[4*i+1] = i
    palette[4*i+2] = i
    palette[4*i+3] = 0xff
```

Make sure the byte order is correct

```python
entry = struct.unpack('I',palette[16:20])
print("palette entry: 0x{:08x}".format(entry[0]))
```

```python
print("The first ten entries in the palette:")
for i in range(10):
    print("{:d}: 0x{:02x}".format(i,palette[4*i+3]),end="")
    print("{:02x}".format(palette[4*i+2]),end="")
    print("{:02x}".format(palette[4*i+1]),end="")
    print("{:02x}".format(palette[4*i]))
print("Length of the palette in bytes: ",len(palette))
```

### Read the pixel data from the file

```python
filename = "capture.dat"
```

```python
captureFile = open(filename,"rb")
pixel_data = captureFile.read()
captureFile.close()
```

```python
print("Size of pixel array: ",len(pixel_data))
print("Should be 96*96 = ",96*96)
```

The image is upside down, so we must revert the data

```python
reverted_pixel_data = bytearray(96*96)
for row in range(96):
    # print("original: {:d}, new: {:d}".format((96-row-1)*96, 96*row))
    for column in range(96):
        reverted_pixel_data[(96-row-1)*96 + column] = pixel_data[96*row + column]
```

### Create the bmp header 

he bmp header looks like this:

| signification | value | offset | size in bytes | 
| --- | --- | --- | --- |
| magic word | 'BM' | 0 | 2 |  
| header size | 14 | 2 | 4 |  
| reserved | 0 | 6 | 2 |  
| reserved | 0 | 8 | 2 |  
| offset to the pixel data |   | 10 | 4 |  

Integers can only be read with 32 bit alignement. Therefore I first read only the magic word 'BM' and then then rest of the header.


```python
fileHdrSize = 14
magic = 'BM'.encode('ascii')
fileSize = fileHdrSize + len(infoHdr) + len(palette) + len(pixel_data)
reserved = 0
offset = fileHdrSize + len(infoHdr) + len(palette)
hdr = magic + struct.pack('ihhi',fileSize,reserved,reserved,offset)
```

```python
printByteArray(hdr)
print("Header length: ",len(hdr))
```

```python
print("The info header: ")
printByteArray(infoHdr)
```

```python
print("file size: ",fileSize)
print("offset:    ",offset)
```

Concatenate hdr, infohdr, palette and pixel_data

```python
print("hdr type: ",type(hdr))
print("infoHdr type: ",type(infoHdr))
print("palette type: ",type(palette))
      
```

```python
bmpData = hdr + infoHdr + palette + reverted_pixel_data
```

```python
print("Length of hdr: ",len(hdr))
print("Length of infoHdr: ",len(infoHdr))
print("Length of palette: ",len(palette))
print("Length of pixel data: ",len(pixel_data))
print("Total length: ",len(bmpData)) 
```

```python
filename = "capture.bmp"
grayscaleBmpFile = open(filename,"wb")
grayscaleBmpFile.write(bmpData)
grayscaleBmpFile.close()
```

Show the constructed image

```python
img = Image.open(filename)
display(img)
```
