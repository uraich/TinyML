## Create a bmp file from a pixel buffer

```python
import struct
```

```python
BI_RGB = 0 # no compression
```

```python
pixelFileName = "uli.dat"
pixelFile = open(pixelFileName,"rb")
pixelBuf = pixelFile.read()
pixelFile.close()
```

```python
print("Length of pixel buffer: ",len(pixelBuf))
print("Type of pixel array: ",type(pixelBuf))
```

Create the bmp header 

```python
fileHdr = bytearray(14)
magic = 'BM'.encode('ascii')
print(magic)
size = 9216
reserved = 0
offset = 500
```

```python
def printByteArray(buffer):
    for i in range(len(buffer) -1):
        print("0x{:02x}, ".format(fileHdr[i]))
    print("0x{:02x}".format(fileHdr[len(buffer) -1]))
```

```python
struct.pack_into('2s',fileHdr,0,magic)
printByteArray(fileHdr)
```
