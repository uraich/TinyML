# Calculate the stroke from the orientation data

Read the stroke data written by the esp32

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stroke_file = open("data/zero_v2_2.txt","r")
stroke_data = stroke_file.read()
stroke_file.close()
```

```python
print(stroke_data)
```

```python
lines = stroke_data.split("\n")
print(lines[0])
```

```python
orientation = []
index = 1
while not "gravity" in lines[index]:
    orientation.append(lines[index].split(','))
    index += 1
print("Index of 'gravitation:'", index)
orientation = np.array(orientation).astype(np.float32)

gravity = []
for index in range (index+1,len(lines)-1):
    gravity.append(lines[index].split(','))
gravity = np.array(gravity).astype(np.float32)
```

```python
print("Orientation")
for i in range(len(orientation)):
    print("{:6.4f},{:6.4f},{:6.4f}".format(orientation[i][0],orientation[i][1],orientation[i][2]))
```

```python
gravity = gravity[:-1]
print("Gravity")
for i in range(len(gravity)):
    print("{:6.4f},{:6.4f},{:6.4f}".format(gravity[i][0],gravity[i][1],gravity[i][2]))
```

```python
current_gravity = gravity[len(gravity)-1]
print("{:6.4f},{:6.4f},{:6.4f}".format(current_gravity[0],current_gravity[1],current_gravity[2]))
```

```python
plt.title("Orientation during gesture")
plt.plot(orientation)
```

```python
plt.title("Gravity during gesture")
plt.plot(gravity)
```

```python
import math
gy = current_gravity[1]
gz = current_gravity[2]
gmag = math.sqrt((gy * gy) + (gz * gz))
if gmag < 0.0001:
    gmag = 0.0001
print("gmag: {:6.4f}".format(gmag))
```

```python
ngy = gy / gmag
ngz = gz / gmag

xaxisz = -ngz
xaxisy = -ngy

yaxisz = ngy
yaxisy = -ngz

print("ngy: {:6.4f}, ngz: {:6.4f}".format(ngy,ngz))
print("xaxisz: {:6.4f}, xaxisy: {:6.4f}, yaxisz: {:6.4f}, yaxisy: {:6.4f}".format(xaxisz,xaxisy,yaxisz,yaxisy))
```

```python
angle_range = 90.0
```

calculate mean orientation

```python
mean_orientation = np.empty((3,),dtype=np.float32)
for i in range(len(orientation)):
    mean_orientation += orientation[i]
mean_orientation /= len(orientation)
print("mean orientation: {:6.4f},{:6.4f},{:6.4f}".format(mean_orientation[0],mean_orientation[1],mean_orientation[2]))
```

```python
n = []
for i in range(len(orientation)):
    n.append((orientation[i] - mean_orientation)/angle_range)
    
```

```python
plt.plot(n)
```

```python
x_axis = []
y_axis = []
for i in range(len(orientation)):
    x_axis.append(xaxisz * n[i][2] + xaxisy * n[i][0]) # n[0]: orientation in x, n[1]: y, n[2]: z
    y_axis.append(yaxisz * n[i][2] + yaxisy * n[i][0])
```

```python
plt.plot(x_axis)
```

```python
plt.plot(y_axis)
```

```python
plt.xlim(-0.6,0.6)
plt.ylim(-0.6,0.6)
plt.plot(x_axis[0],y_axis[0],"ro")
plt.plot(x_axis,y_axis)
```
