# Plot input tensors of the magic wand example

```python
import numpy as np
import matplotlib.pyplot as plt
```

Read the pixel values

```python
f = open("0.bin","rb")
pixels = np.frombuffer(f.read(),dtype=np.uint8).reshape((32,32,3))
f.close()
```

```python
plt.imshow(pixels)
```

```python
stroke_pixels = []
for i in range(10):
    f = open("{:d}.bin".format(i),"rb")
    stroke_pixels.append(np.frombuffer(f.read(),dtype=np.uint8).reshape((32,32,3)))
```

```python
print(len(stroke_pixels))
```

Now that we know how to plot a single stroke, we can extend the plot to all 10 digit strokes

```python
fig,ax = plt.subplots(2,5)
fig.suptitle("Stroke test data for all 10 digits")
for i in range(2):
    for j in range(5):
        ax[i,j].set_axis_off()
        ax[i,j].imshow(stroke_pixels[i*5+j])
```
