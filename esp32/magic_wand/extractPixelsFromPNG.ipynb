# Extract pixels from png image

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = "images/7/2.png"
```

```python
im = Image.open(filename)
im.convert('RGB')
imageSizeW, imageSizeH = im.size
pixels = np.array(im).reshape((32,32,3))
```

```python
plt.imshow(pixels)
```
