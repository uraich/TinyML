## Calculate BLE characteristics for strokes

```python
stroke_transmit_stride = 2
stroke_transmit_max_length = 160
```

```python
stroke_max_length = stroke_transmit_max_length * stroke_transmit_stride * 1 * stroke_transmit_max_length
stroke_points_byte_count = 2 * 4 * stroke_transmit_max_length
stroke_struct_byte_count = (2 * 4 ) + stroke_points_byte_count
print("stroke_max_length: ",stroke_max_length)
print("stroke_struct_byte_count: ",stroke_struct_byte_count)
```
